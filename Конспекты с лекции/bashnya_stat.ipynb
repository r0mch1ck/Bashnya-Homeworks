# Предобработка данных

### Смотрим статистику по тем, кто не указал кафедру

```python
df['Кафедра вашего обучения'].isna().value_counts()
```

### Заменяем nan на текст

```python
df.fillna('Кафедра не указана', inplace=True)
df
```

### Кодируем результаты опроса в число

```python
df.replace('Не владею', 0, inplace=True)
df.replace('Слышал, но не использовал', 0, inplace=True)
df.replace('Владею недостаточно', 1, inplace=True)
df.replace('Владею', 2, inplace=True)
df
```

### Добавляем новую метрику

```python
df['Уровень знаний'] = df.iloc[:, 5:].sum(axis=1)
df
```

# Приводим факультеты к единому стандарту

```python
faculties = ['АК','БМТ','ИБМ','ИУ','Л','МТ', 'ПС','РК','РКТ','РТ','СГН','СМ','ФН','Э','ЮР']
for index, row in df.iterrows():
    for i in faculties:
        if i.lower() in row['Кафедра вашего обучения'].lower():
            df.loc[index, 'Кафедра вашего обучения'] = i
df
```

### Для людей из другого вуза приводим к стандартному виду кафедру

```python
for index, row in df.iterrows():
    if row['Кафедра вашего обучения'] not in faculties and (row['Кафедра вашего обучения'] != 'Кафедра не указана' or row['Ваш ВУЗ'] != 'МГТУ им Баумана'):
        df.loc[index, 'Кафедра вашего обучения'] = 'Не МГТУ'
```

## Нужно избавлятся от выбросов

```python
df = df[~df['Имя и Фамилия'].str.contains('Иван Иванов', na=False)]
```

```python
x = np.linspace(-5, 5, 1000)
normal_dist = (1 / (np.sqrt(2 * np.pi))) * np.exp(-0.5 * x**2)

non_normal_dist = np.exp(-x) * (x > 0)

plt.figure(figsize=(10, 6))
plt.plot(x, normal_dist, label='Нормальное распределение', color='blue')
plt.plot(x, non_normal_dist, label='Ненормальное распределение', color='red')

plt.title('Нормальное распределение vs Ненормальное')
# plt.axhline(0, color='black', linewidth=0.5)
# plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

plt.grid(True)
plt.show()
```

```python
def sort(arr):
    arr.sort()
    left, right = [], []
    for i in range(len(arr)):
        if i % 2 == 0:
            right.append(arr[-(i+1)])
        else:
            left.append(arr[-(i+1)])
    return left[::-1] + right

df_copy = df[~df['Имя и Фамилия'].str.contains('Иван Иванов', na=False)].copy()

sorted_levels = sort(df_copy['Уровень знаний'].tolist())
df_copy['Уровень знаний'] = sorted_levels

mu = df_copy['Уровень знаний'].mean()
sigma = df_copy['Уровень знаний'].std()

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / (1.5 * sigma))**2)

plt.figure(figsize=(10, 6))
plt.bar(range(len(df_copy)), df_copy['Уровень знаний'], alpha=0.6, label='Уровень знаний')
x_gauss = np.linspace(0, len(df_copy) - 1, 1000)
plt.plot(x_gauss, y * max(df_copy['Уровень знаний']) * 8.0, color='red', label='Гауссово распределение')
plt.ylabel('Уровень знаний')
plt.title('Распределение уровня знаний и Гауссово распределение')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
mean_total = df['Уровень знаний'].mean()
std_total = df['Уровень знаний'].std()
mean_total, std_total
```

### Фильтрация DataFrame по правилу трех сигм

```python
lower_bound = mean_total - 3 * std_total
upper_bound = mean_total + 3 * std_total

df = df[(df['Уровень знаний'] >= lower_bound) & (df['Уровень знаний'] <= upper_bound)]
```

# Строим колличество заявок от времени

### Приводим колонку время к типу date time

```python
df['Время создания'] = pd.to_datetime(df['Время создания'])
```

### Создание DataFrame для хранения количества заявок по полу часам, объединение данных с количеством заявок

```python
start_time = '2024-10-05 20:30:00'
end_time = '2024-10-05 22:30:00'

time_range = pd.date_range(start=start_time, end=end_time, freq='15T')

# Создание DataFrame для хранения количества заявок по полчасам
requests_per_half_hour = pd.DataFrame({'15 минут': time_range})

# Объединяем данные с количеством заявок
requests_per_half_hour = requests_per_half_hour.merge(
    df.groupby(df['Время создания'].dt.floor('15T')).size().reset_index(name='Количество заявок'),
    left_on='15 минут',
    right_on='Время создания',
    how='left'
).fillna(0)  # Заполняем NaN нулями для промежутков без заявок
requests_per_half_hour
```

### Построение графика зависимости количества заявок от времени

```python
plt.figure(figsize=(12, 6))
plt.plot(requests_per_half_hour['15 минут'], requests_per_half_hour['Количество заявок'], marker='o', linestyle='-.')
plt.title('Количество заявок по времени', fontsize=14)
plt.xlabel('Время', fontsize=12)
plt.ylabel('Количество заявок', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
